```python
#importing necessary libraries
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
import re
#!pip install urlextract
from urlextract import URLExtract
```

```python
#Some information that will explain why I did what I did to get the data from the website:
#the url-address will not change as we navigate from one page to another (there are 72 pages containing character tabs; and each tab on the page is a separate page).
#The url-address is constantly at https://www.marvel.com/characters for all of the pages containing tabs with characters.
#First 12 characters on each page are the main characters from movies and they are present on each page of 72 pages with characters. 
#To extract all of the urls of the character tabs that I need (I needed all of them since there is no page that holds all of the information about the characters, it is all on separate pages)
#after trying a few things I decided to use the sitemap webpages that hold all the urls for the tabs with characters information. The urls from the sitemap are in the list bellow
```

```python
#list with sitemap urls that contain character urls
urlslist=["https://www.marvel.com/sitemap-7.xml", "https://www.marvel.com/sitemap-8.xml","https://www.marvel.com/sitemap-9.xml"]
```

```python
#scraping urls with characters from the sitemap pages of the website marvel.com and adding them to a list
#all of these urls will be used to extract the necessary information
urls=[]
for i in urlslist:
    url = i
    web_stuff = requests.get(url) # reading the url
    html = web_stuff.text # returning the page's content in unicode
    soup = BeautifulSoup(html, "html.parser") # getting the page's html
    new_soup=soup.find_all("url")
    soup1=str(new_soup)
    extractor = URLExtract()
    for url in extractor.gen_urls(soup1):
        if 'characters' in url:
            urls.append(url) 
    print(urls)
```

```python
#checking the number of urls in the list. It matches to what the website search field says
len(urls)
```

```python
#WARNING:THIS TAKES A VERY LONG TIME TO RUN!
#creating lists that will hold the data that I need to create a dataframe
names=[]
description=[]
heights=[]
weights=[]
hair = []
eye=[]
universe=[]
education = []
origin=[]
identity=[]
for i in urls:
    my_url = i
    web_stuff = requests.get(my_url) # reading the url from the urls list
    html = web_stuff.text # returning the page's content in unicode
    soup = BeautifulSoup(html, "html.parser") # getting the page's html
    soup.prettify()
    #retrieving the values from the field description
    try:
        char_description = soup.find("div", class_="text").text
    except Exception as e:
        char_description=None
    #appending the values to the list description
    description.append(char_description)
    #retrieving the name
    try:
        match=soup.find('div', class_='masthead__container masthead__container_playing-false')
        name=match.h1.find(class_="masthead__headline").text
    except Exception as e:
        name=None
    #appending the values to list names
    names.append(name)
    #converting soup to string which will be needed for the future operations
    soup_str = str(soup)
    #since height, weight, eyes and hair have exact same div name and class name find_all lets us find them on the page,
    #but a lot of pages only have 1 or 2 or 3 out of these four variables and since I have been trying to index them
    #value[0]-height;value[1]-weight and etc I keep getting errors when some of these values are missing, or one of the values
    #takes a spot of another value (when one of the four is missing) and the information is incorrect. 
    #If you look at the pages on marvel.com,when "height" is missing "weight" takes its position on the page.
    #I found a different solution where I was able to locate the text preceeding the value with one of the words (height,weight,hair,eyes)
    #and the text following the value, and extract the text that is in between (the one that I need)
    #The same is true for the rest of the variables(universe, identity and etc).
    #They take the position of each other on the page if one of them is missing, and they have the same div and class.
    # If the text I am trying to extract is not present,this value is absent on the page and then I add NONE.
    text = 'height</p><p class="bioheader__stat">'
    if text in soup_str:
        char_height = soup_str.split('height</p><p class="bioheader__stat">', 1)
        character_h = char_height[1].split('</p></div>')[0]
        heights.append(character_h)
    else:
        heights.append(None)
    text1='weight</p><p class="bioheader__stat">'                                 
    if text1 in soup_str:                                 
        char_weight = soup_str.split('weight</p><p class="bioheader__stat">',1)
        character_weight = char_weight[1].split('</p></div></div>')[0]
        weights.append(character_weight)
    else:
        weights.append(None)
    text2 = 'hair</p><p class="bioheader__stat">'
    if text2 in soup_str:
        char_hair = soup_str.split('hair</p><p class="bioheader__stat">',1)
        character_hair = char_hair[1].split('</p></div></div>')[0]
        hair.append(character_hair)
    else:
        hair.append(None)
    text3 = 'eyes</p><p class="bioheader__stat">'
    if text3 in soup_str:
        char_eye = soup_str.split('eyes</p><p class="bioheader__stat">',1)
        character_eye = char_eye[1].split('</')[0]
        eye.append(character_eye)
    else:
        eye.append(None)   
    text4 = 'Universe</p><ul class="railBioLinks"><li>'                         
    if text4 in soup_str:
        char_uni = soup_str.split('Universe</p><ul class="railBioLinks"><li>',1)
        character_uni = char_uni[1].split('</')[0]
        universe.append(character_uni)
    else:
        universe.append(None) 
    text5='Education</p><ul class="railBioLinks"><li>'
    if text5 in soup_str:
        char_edu = soup_str.split('Education</p><ul class="railBioLinks"><li>',1)
        character_edu = char_edu[1].split('</')[0]
        education.append(character_edu)
    else:
        education.append(None)
    text6='Place of Origin</p><ul class="railBioLinks"><li>'
    if text6 in soup_str:
        char_or = soup_str.split('Place of Origin</p><ul class="railBioLinks"><li>',1)
        character_or = char_or[1].split('</')[0]
        origin.append(character_or)
    else:
        origin.append(None)
    text7='l">Identity</p><ul class="railBioLinks"><li>'
    if text7 in soup_str:
        char_iden = soup_str.split('l">Identity</p><ul class="railBioLinks"><li>',1)
        character_iden = char_iden[1].split('</li></ul>')[0]
        identity.append(character_iden)
    else:
        identity.append(None)
#printing this value just to check the number of items per list. It is the same as the number of urlswith characters information.
print(len(heights) )  
```

```python
import pandas as pd
#n creating a data frame with columns:name,description,heights,weights,hair,eye,universe,education,origin,identity 
df = pd.DataFrame()
df['name']=names
df['description']=description
df['heights']=heights
df['weights']=weights
df['hair']=hair
df['eye']=eye
df['universe']=universe
df['education']=education
df['origin']=origin
df['identity']=identity
df.tail()
```

```python
#converting feet and inches into decimals for possible operations with numbers in the future
import re
def decimalize_feet(s):
    m = re.match(r'^(\d+)\'(\d+)\"$', s)
    if m:
        return round(float(m.group(1)) + float(m.group(2)) / 12.,2)
    else:
        return None
newdata.loc[:,'height'] = newdata.loc[:, 'height'].astype(str)
newdata['height']=newdata['height'].apply(lambda x: decimalize_feet(x))
newdata.head(10)
```
