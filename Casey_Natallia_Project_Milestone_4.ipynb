```python
#importing the libraries needed to pull the needed data from the ComicVine Api and loading it into a data frame.
import urllib.request, urllib.parse
from urllib.error import HTTPError,URLError
import pandas as pd
import json
offset = 0
all_items = []

# looping through all offset by 100 and returning JSON object
#only 100 observations per call are allowed; there are 137187 total. 
#unfortunately filtering by publisher at the moment is not supported (I only need Marvel characters):
#my decision was to load all of the characters JSON data, create a dataframe and then filter it by a URL that is used for Marvel
#the final step was to concatenate all of the dataframes into one (since only 100 observations could be processed at a time,
#I ended up having multiple dataframes that were concatenated successfully into one.)
for offset in range(0, 137187,100):

    url = "http://comicvine.com/api/characters/?api_key=b1af0ab4593ece4e9f218e1fecf953cca089d31f&format=json&offset=%d" % offset           
    uh = urllib.request.urlopen(url)
    #decoding data    
    data = uh.read().decode()
    x = json.loads(data) 
    y = x['results']
    df = pd.read_json(json.dumps(y))
    #only adding the data for the Marvel publisher
    newdf = df.loc[df['publisher'] == {'api_detail_url': 'https://comicvine.gamespot.com/api/publisher/4010-31/', 'id': 31, 'name': 'Marvel'}]
    #appending dataframe with only Marvel publisher data to a list 
    all_items.append(newdf)  
       
    print(offset)
 
 #concatinating the resulting dataframes into one
result = pd.concat(all_items, ignore_index=True)
```

```python
#saving the data frame to a csv file
result.to_csv('comicvine.csv')
```
